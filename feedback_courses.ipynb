```python


java_full_stack_df = df[df['Name of Course'] == 'Java Full Stack']


bar_palette = 'Set1'

plt.figure(figsize=(15, 5))
sns.barplot(
    x='Batch code',
    y='average_rating',
    hue='Name of Trainers',
    data=java_full_stack_df,
    palette=bar_palette,  
    ci=None,
    dodge=True,
    saturation=2,
    linewidth=0.1,
    width=2
)

plt.xlabel('Batch Code')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Each Trainer in Java Full Stack Course')


legend_labels = java_full_stack_df['Name of Trainers'].unique()
legend_colors = sns.color_palette(bar_palette, n_colors=len(legend_labels))

custom_legend = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label)
    for label, color in zip(legend_labels, legend_colors)
]

plt.legend(handles=custom_legend, title='Trainer Name', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()  
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.savefig('Average Ratings for Each Trainer in Java Full Stack Course.png')

```

```python
#same visulization in heatmap
import seaborn as sns
import matplotlib.pyplot as plt


java_full_stack_df = df[df['Name of Course'] == 'Java Full Stack']


heatmap_data = java_full_stack_df.pivot_table(index='Name of Trainers', columns='Batch code', values='average_rating')

plt.figure(figsize=(15, 5))
sns.heatmap(
    heatmap_data,
    annot=True,  
    fmt=".2f",   
    cmap="YlGnBu",  
    cbar_kws={'label': 'Average Rating'}  
)

plt.xlabel('Batch Code')
plt.ylabel('Trainer Name')
plt.title('Average Ratings for Each Trainer in Java Full Stack Course')
plt.yticks(fontsize=20)
plt.xticks(fontsize=12)
plt.savefig('Average Ratings for Each Trainer in Java Full Stack Course.png')

plt.show()
```
